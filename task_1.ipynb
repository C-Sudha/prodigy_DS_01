```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('student-por.csv')
```

```python
df
```

```python
df.head(5)
```

```python
ages = df['age']
```

```python
ages
```

```python
# Create a histogram
plt.hist(ages, bins=10, color='blue', edgecolor='black')

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Distribution of Ages')
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Show the plot
plt.show()
```

```python
gender_counts = df['sex'].value_counts()
```

```python
gender_counts
```

```python
# Create a bar chart
gender_counts.plot(kind='bar', color='orange', edgecolor='black')

# Add labels and title
plt.xlabel('Gender')
plt.ylabel('Count')
plt.title('Distribution of Genders')

# Show the plot
plt.show()
```
